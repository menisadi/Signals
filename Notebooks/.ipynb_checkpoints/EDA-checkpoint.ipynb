```python
from sklearn.metrics.pairwise import haversine_distances
```

```python
signals_locations = signals_data[['Latitude', 'Longitude']]
antennas_locations = antennas_data[['Latitude', 'Longitude']]

distances = haversine_distances(signals_locations.values, antennas_locations.values)
```

```python
distances.shape
```

```python
closest_antenna_indices = np.argmin(distances, axis=1)
```

```python
closest_antenna_indices.shape
```

```python
closest_antenna_types = antennas_data.iloc[closest_antenna_indices]['type'].values
```

```python
signals_data['closest_antenna_type'] = closest_antenna_types
signals_data['distance_to_closest_antenna'] = np.min(distances, axis=1)*1000
```

```python
signals_data['distance_to_closest_antenna'].plot(kind='hist', bins=30)
```

## Learning

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

### First Attempt - Linear Regression

```python
X = signals_data[['closest_antenna_type', 'distance_to_closest_antenna']]
y = signals_data['signal']
```

```python
X = pd.get_dummies(X, columns=['closest_antenna_type'], drop_first=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
test_score = lr.score(X_test, y_test)
print("Test score: ", test_score)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, mean_squared_error
```

```python
predictions = lr.predict(X_test)
```

```python
mean_squared_error(y_test, predictions)
```

### Second Attempt - KNN

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
X = signals_data[['Latitude', 'Longitude']]
y = signals_data['signal']
```

```python
# X = pd.get_dummies(X, columns=['closest_antenna_type'], drop_first=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
knn = KNeighborsRegressor()
knn.fit(X_train, y_train)
```

```python
predictions = knn.predict(X_test)
```

```python
mean_squared_error(y_test, predictions)
```

```python
# Initialize a list to store the mean squared errors for different values of k
mse = []

# Try k values from 1 to 10
for k in range(1, 11):
    # Create a KNN model with the current value of k
    knn = KNeighborsRegressor(n_neighbors=k)
    # Fit the model to the training data
    knn.fit(X_train, y_train)
    # Use the model to make predictions on the test set
    y_pred = knn.predict(X_test)
    # Calculate the mean squared error of the predictions
    mse.append(mean_squared_error(y_test, y_pred))
```

```python
# Plot the mean squared error for each value of k
plt.plot(range(1, 11), mse)
plt.xlabel('k')
plt.ylabel('Mean Squared Error')
plt.show()
```

```python
knn = KNeighborsRegressor(n_neighbors=2)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
# Calculate the mean squared error of the predictions
mean_squared_error(y_test, y_pred)
```

### Back to Fueature Engineering 

```python
types = antennas_data.type.unique()
```

```python
type_distances = distances[:,(antennas_data['type']==types[0]).to_numpy()]
```

```python
# type_distances.min()
```

```python
closest_antenna_of_type = np.argmin(type_distances, axis=1)
```

```python
closest_antenna_of_type
```

```python
for type in types:
    type_distances = distances[:,(antennas_data['type']==type).to_numpy()]
    distance_to_closest_antenna_of_type =  np.min(type_distances, axis=1)*1000
    signals_data[f'closest_{type}'] =  distance_to_closest_antenna_of_type
```

```python
signals_data.head()
```

```python
X = signals_data[['closest_antenna_type','closest_D','closest_A','closest_C','closest_B']]
y = signals_data['signal']
```

## Neural Network?

```python
from sklearn.preprocessing import MinMaxScaler

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation,Dropout,BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping

from tensorflow.keras.layers import Dropout
```

```python
model = Sequential([
    Dense(units=7, activation='relu'),
    Dropout(0.2),
    BatchNormalization(),
    Dense(units=14, activation='relu'),
    Dropout(0.2),
    BatchNormalization(),
    Dense(units=28, activation='relu'),
    Dropout(0.2),
    BatchNormalization(),
    Dense(units=7, activation='relu'),
    Dropout(0.2),
    BatchNormalization(),
    Dense(1),
])
```

```python
model.compile(
    optimizer='adam',
    loss='mse',
)
```

```python
early_stop = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=25)
```

```python
history = model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=500,
    verbose=1,
    callbacks=[early_stop]
)
```

```python
model_loss = pd.DataFrame(model.history.history)
model_loss.plot()
```

### Well, Ridge maybe?

```python
from sklearn.linear_model import Ridge
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
rd = Ridge()
rd.fit(X_train, y_train)
```

```python
test_score = rd.score(X_test, y_test)
print("Test score: ", test_score)
```

```python
predictions = rd.predict(X_test)
```

```python
mean_squared_error(y_test, predictions)
```

```python
mse = []

for a in range(1, 150):
    rd = Ridge(alpha=a/2)

    rd.fit(X_train, y_train)
    y_pred = rd.predict(X_test)
    mse.append(mean_squared_error(y_test, y_pred))
```

```python
plt.plot(range(1, 150), mse)
plt.xlabel('a')
plt.ylabel('Mean Squared Error')
plt.show()
```
